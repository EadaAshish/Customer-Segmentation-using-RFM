### Creating RFM Scores

```python
# Using 3 Quantile to categorize in
# 1. Champions
# 2. Potential Customers
# 3. Need Attention

quantile = RFM.quantile(q=[0.33,.66])
quantile = quantile.to_dict()
quantile
```

```python
# Function to create R, F and M segements

# Smaller Recency Value is Better
def R_Score(val,col,df):
    if val <= df[col][0.33]:
        return 1
    elif val <= df[col][0.66]:
        return 2
    else:
        return 3

# Greater Frequency and Monetory Values are better
def F_M_Score(val,col,df):
    if val <= df[col][0.33]:
        return 3
    elif val <= df[col][0.66]:
        return 2
    else:
        return 1
```

```python
RFM['R_Score'] = RFM['Recency'].apply(R_Score, args=('Recency',quantile))
RFM['F_Score'] = RFM['Frequency'].apply(F_M_Score, args=('Frequency',quantile))
RFM['M_Score'] = RFM['Monetary'].apply(F_M_Score, args=('Monetary',quantile))

RFM
```

```python
# Calculate the concatenated value for the group code
RFM['RFM_Group'] = RFM.R_Score.map(str) + RFM.F_Score.map(str) + RFM.M_Score.map(str)

# Calculate the total score of RFM Features
RFM['RFM_Score'] = RFM[['R_Score', 'F_Score', 'M_Score']].sum(axis=1)

RFM
```

```python
# Validating the RFM Score
RFM[RFM['RFM_Score'] == 4]
```

## Assigning Categories

```python
Categories = ['Champions', 'Potential Customers', 'Need Attention']
Score_cuts = pd.qcut(RFM.RFM_Score, q=3, labels=Categories)
RFM['Category'] = Score_cuts.values
RFM
```

```python
# Validating Data for RFM_Score & Category
RFM[RFM['RFM_Group'] == '122'].sort_values('Monetary', ascending=False)
```
